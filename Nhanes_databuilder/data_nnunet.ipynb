```python
import os
import shutil
import random

# Define the folder path
home = "/home/soumya/yuanhan/Dataset800_Nhanes/"
```

```python
iTr = home + "imagesTr"
iTs = home + "imagesTs"
lTr = home + "labelsTr"
lTs = home + "labelsTs"

folders = [iTr, iTs, lTr, lTs]

for folder_path in folders:
    countC = 0
    countL = 0
    count = 0

    print(folder_path)
    for filename in os.listdir(folder_path):
        if filename.startswith('C'):
            countC += 1
        if filename.startswith('L'):
            countL += 1
        count += 1

    print("Number of files starting with 'C':", countC)
    print("Number of files starting with 'L':", countL)
    print("Total number of files:", count)
```

```python
#### Renaming Images
```

```python
def rename_files(directory):
    count = 0
    for filename in os.listdir(directory):

        count = count + 1

        if filename.endswith(".png"):
            parts = filename.split("_")
            new_filename = f"NHANES_{count:03d}_0000.png"
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))

# Replace 'directory_path' with the path to the directory containing the files
directory_path = home + "imagesTr"  # Update this with the correct path
rename_files(directory_path)

directory_path = home + "imagesTs"  # Update this with the correct path
rename_files(directory_path)
```

```python
#### Renaming Labels
```

```python
def rename_files(directory):
    count = 0
    for filename in os.listdir(directory):

        count = count + 1

        if filename.endswith(".png"):
            parts = filename.split("_")
            new_filename = f"NHANES_{count:03d}.png"
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))

# Replace 'directory_path' with the path to the directory containing the files
directory_path = home + "labelsTr"  # Update this with the correct path
rename_files(directory_path)

directory_path = home + "labelsTs"  # Update this with the correct path
rename_files(directory_path)
```
