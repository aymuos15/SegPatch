```python
from collections import defaultdict
import numpy as np
import cv2

image_save_path = '/home/soumya/oxford/Dataset900_Vindr/imagesTr/'
label_save_path = '/home/soumya/oxford/Dataset900_Vindr/labelsTr/'

# image_save_path = '/home/soumya/oxford/final/imagesTs/'
# label_save_path = '/home/soumya/oxford/final/labelsTs/'

# Function to plot DICOM image with bounding box for multiple annotations
def add_bounding_boxes_to_label(image_annotations, id, dicom_path=dicom_path_train):
    # Load DICOM image
    dicom_file = dicom_path + f"{image_annotations[0]['image_id']}.dicom"  # Assuming all annotations belong to the same image
    dicom_data = pydicom.dcmread(dicom_file)

    label = np.zeros((dicom_data.pixel_array.shape[0], dicom_data.pixel_array.shape[1]))
    
    # Add bounding boxes to label
    for annotation in image_annotations:
        xmin = int(annotation['xmin'])
        ymin = int(annotation['ymin'])
        xmax = int(annotation['xmax'])
        ymax = int(annotation['ymax'])
        
        # Set region within bounding box to 1
        label[ymin:ymax, xmin:xmax] = 1

    # to work with nnunet
    cv2.imwrite(image_save_path + f'{id}.png', dicom_data.pixel_array)
    cv2.imwrite(label_save_path + f'{id}.png', label)

# Group annotations by image ID
annotations_by_image = defaultdict(list)
for index, row in df.iterrows():
    annotations_by_image[row['image_id']].append(row)

# Iterate over grouped annotations and plot each DICOM image with bounding boxes
# count = 0
for image_id, annotations in annotations_by_image.items():
    print(f"ID: {image_id} label and raw saved.")
    add_bounding_boxes_to_label(annotations, image_id)
    # count += 1
    # if count == 15:
    #     break
```
